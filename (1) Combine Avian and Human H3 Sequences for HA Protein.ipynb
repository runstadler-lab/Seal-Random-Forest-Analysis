```python
from Bio import SeqIO, SeqRecord
```

```python
human_sequences = SeqIO.parse('Human H3 Sequences.fasta', 'fasta')
human_sequences
```

```python
combined_sequence = []
```

```python
for hs in human_sequences:
    combined_sequence.append(hs)
```

```python
combined_sequence
```

```python
avian_sequences = SeqIO.parse('Avian H3 Sequences.fasta', 'fasta')
avian_sequences
```

```python
for avs in avian_sequences:
    combined_sequence.append(avs)
```

```python
output_handle = open('Human and Avian H3 Sequences.fasta', 'w+')
SeqIO.write(combined_sequence, output_handle, 'fasta')
```
